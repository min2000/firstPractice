# DataFrame column간 상관관계 계산하기_15강

**변수(column) 사이의 상관계수(correlation)**
~~~
corr 함수를 통해 상관계수 연산 (-1,1 사이의 결과)

# DataFrame NaN 데이터 처리

**NaN 값 확인**
~~~
info 함수를 통해 개수 확인
insa 함수를 통해 boolean 타입으로 확인
~~~

```python
stock_data.info()    #모든 data가 1232개로 누실이 없다
```

```python
stock_data.isna()    #True인 경우에는 값이 NaN이다. False만 있으므로 누락 data 없음
```

```python
stock_data['Date'].isna()   #특정한 column의 누실이 있는지 확인도 가능
```

**NaN 처리 방법**
~~~
데이터에서 삭제
    dropna 함수
다른 함수로 치환
    fillna 함수
    ~~~

NaN 데이터 삭제하기

```python
stock_data.dropna()   #NaN이 있는 row 삭제, NaN이 하나라고 있으면 row 전체 삭제
```

```python
stock_data.dropna(subset=['Open','Colse'])   #subset을 사용하면 Open이라는 column에 NaN이 있어도 삭제되지 않음
```

```python
stock_data.dropna(axis=1)      #column 중 하나라도 NaN이면 column 자체를 지우기
```

**NaN 값 대체하기**
~~~
평균으로 대체하기
특정 column의 평균 구하기
~~~

```python
stock_data['Expection'].fillna(stock_data['Expextion'].mean())   #Expection중 NaN이 있으면 평균값으로 대체
```

```python
#High가 45 이상인 data의 Low 평균
mean1=stock_data[stock_data['High']>45]['Low'].mean()
```

```python
mean1
```

```python
stock_data[stock_data['High']>45]['Low'].fillna(mean1)
```

```python
#변화된 것을 복사본이 아니라 실제 데이터에 적용하기

stock_data.loc[stock_data['High']>45,'Low']=stock_data[stock_data['High']>45]['Low'].fillna(mean1)
```
