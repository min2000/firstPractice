# DataFrame 원하는 row(데이터)만 선택하기

**dataframe slicing**
~~~
dataframe의 경우 기본적으로 [ ] 연산자가 column 선택에 사용
하지만 sclicing은 row 레벨로 지원
~~~

```python
stock_data[:10]   #indexing은 안되지만 sclicing은 숫자를 적어주면 몇개의 줄을 보여줄 건지 row level로 작동
```

```python
import numpy as np
```

**row 선택하기**
~~~
Series의 경우 [ ]로 row 선택이 가능하나, DataFrame의 경우는 기본적으로 column을 선택하도록 설계
.loc, .iloc로 row 선택 가능
    loc: 인덱스 자체를 사용
    iloc: 0 based index로 사용
    이 두 함수는 ,를 사용하여 column 선택도 가능
~~~

```python
stock_data.info()
```

```python
stock_data.index=np.arange(100,1332)    #loc과 iloc의 차의점을 알기 위해 index를 인위적으로 바꾸기
```

```python
stock_data.head()
```

```python
stock_data.loc[986]
```

```python
stock_data.loc[[100,991,1200]]
```

```python
stock_data.iloc[0]
```

**row, column 동시에 선택하기**
~~~
loc, iloc 속성을 이용할 때, 콤마를 이용하요 둘 다 명시 가능
~~~

```python
stock_data.loc[[991,1020,110,990],['High','Low']]    #row 명시 이후에 column 명시 가능
```

```python
stock_data.iloc[[991,0,110,100],[2,3]]    #iloc은 index를 number로 표시   
```

**Boolean Selection으로 row 선택하기**
~~~
numpy에서와 동일한 방식으로 해당 조건에 맞는 row만 선택
~~~

~~~
High가 44.5 이상이면서 Low는 43.5 이상인 row만 선택
~~~

```python
High_=stock_data['High']>44.5
Low_=stock_data['Low']>43.5

High_
Low_
```

```python
stock_data[High_&Low_]    #index와 맞는 boolean list 생성 후 [ ]안에 넣기
```
