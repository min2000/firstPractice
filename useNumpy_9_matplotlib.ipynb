**연습문제**
~~~
로또 번호 자동 생성기(함수로)를 만드시오 (random으로 6개 뽑기)
numpy를 이용하요 pi(원주율) 값을 계산하시오
    *몬테 카를로 방법 이용하기 (pi값 계산): 난수를 발생시켜 근사값을 발생
    ~~~

```python
import numpy as np
```

```python
def generate_lotto_nums():
    return np.random.choice(np.arange(1,46),size=6, replace=False)   #replace=False가 비복원 추출
    
generate_lotto_nums()
```

pi/4 : 1 = (4분원 안에 생성된 점 개수) : 전체 시도 횟수

pi=4*(4분원 안에 생성된 점 개수)/(전체 시도 횟수)
(해당 면적)/(전체 면적)=만족하는 사건이 일어날 확률

```python
total=int(1e7)   #1e7는 천만개 의미 10의 7승
points=np.random.rand(total,2)    
print(points)
points.shape   #(1000,2)  #1000만 개의 숫자가 
```

```python
np.sum(points**2,axis=1)    #천만개의 원소를 갖는 1차원 벡터
```

```python
np.sum(points**2,axis=1) <1    #벡터와 스칼라를 비교하면 broadcasting이 발생하면서 True/False 판별
```

```python
np.sum(np.sum(points**2,axis=1) <1 )
```

```python
4*np.sum(np.sum(points**2,axis=1) <1 )/total
```
