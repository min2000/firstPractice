# 숫자형 데이터의 범주형 데이터화_18강

**astype 사용하여 간단하게 타입만 변화**
~~~
숫자형 type을 범주형으로 변화
~~~

```python
stock_data.info()
```

```python
stock_data['Volume']=stock_data['Volume'].astype(str)   #type를 바꾸는데 사용
```

```python
stock_data.info()
```

**변환 로직을 함수로 만들고 apply 함수로 적용**
~~~
숫자형 type을 

```python
import math
```

```python
def open_catagorize(open):
    if math.isnan(open):
        return -1
    return math.floor(open/2)*2   #(open/10)*10는 일의 자리수를 삭제함
```

```python
stock_data['Open'].apply(open_catagorize)   #(open/2)*2 는 모두 짝수로 변환
```

# 범주형 데이터 잔처리 하기 (one-hot encoding)

**One-hot encoding**
~~~
범주형 데이터는 분석단계에서 계산이 어렵기 때문에 숫자형으로 변경이 필요
범주형 데이터의 각 번주(catagory)를 column 레벨로 변경
해당 범주에 포함되면, 1아니면 0으로 채우는 인코딩 기법
pandas.get_dummies 함수 사용
    drop_first: 첫번째 카테고리 값은 사용하지 않음
    ~~~

```python
pd.get_dummies(stock_data)
```

```python
pd.get_dummies(stock_data,columns=['High','Low'])   #각 High와 Low값이 column level로 올라간다
```

```python
pd.get_dummies(stock_data,columns=['High','Low'],drop_first=True)   #하나의 column을 없앰
                                                     #나머지의 정보들로 없어진 정보 충분히 유추 가능
```
