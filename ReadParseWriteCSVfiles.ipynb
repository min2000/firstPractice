```python
with open('train.csv','r') as csv_file:
    csv_reader=csv.DictReader(csv_file)   #dictionary 형태로 바꾼다
    
    with open('train_practice.csv','w') as new_file:
        fieldnames=['key','fare_amount','pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']
        
        csv_writer=csv.DictWriter(new_file, fieldnames=fieldnames, delimiter='\t')
        csv_writer.writeheader()   #header도 작성할 필요가 있을 때 이 코드를 사용
        
        for line in csv_reader:
            del line['pickup_latitude']  #특정 key 지우고 싶을 때 사용
            csv_writer.writerow(line)

#dictionary reader는 아무것도 제공할 필요가 없지만 dictionary writer는 file의 fieㅣd name을 제공해야 한다
```
